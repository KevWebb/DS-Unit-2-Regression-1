```python
import pandas as pd
voteshare=pd.read_csv('https://raw.githubusercontent.com/KevWebb/DS-Unit-2-Regression-1/master/VotingShare.csv')
voteshare
```

```python
unrate = pd.read_csv('C:/Users/Kevona Jean/Desktop/DS4 Notebooks/Unit 2 Sprint 1 Regression/unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate.index = unrate['DATE']
unrate.head()
```

```python
unrate['Year'] = unrate['DATE'].apply(lambda x: x.year)
unrate.head()
```

```python
df = pd.merge(voteshare, unrate, on='Year')
df.drop(columns= ['Unnamed: 0', 'DATE'], inplace=True)
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

features = ['UNRATE', 'Average Recent Growth in Personal Incomes']
target = 'Incumbent Party Vote Share'
X = df[features]
y= df[target]
```

```python
model.fit(X,y)
```

```python
y_pred= model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, y_pred)
```

```python
df['Linear Regression, Unrate'] = y_pred
df['Error'] = y_pred - df[target]
df['Absolute Error'] = df['Error'].abs()
df
```

Leave-One-Out Cross-Validation

```python
from sklearn.linear_model import LinearRegression
print('Leave-One-Out Cross-Validation, Do-It-Yourself implemnetation')

maes =[]

for year in df['Year']: 
    print(f'Fit Linear Regression on all years except {year}')
    
    #Choose model hyperparmeters by instantiating this class
    model = LinearRegression()
    
    #Arrange data into X features matrix and y target 
    train = df[df['Year'] !=year]
    test = df[df['Year'] == year]
    X_train = train[features]
    y_train = train[target]
    X_test = test[features]
    y_test = test[target]
    
    #Fit the model to your data by calling
    model.fit(X_train, y_train)
    
    #Apply the model to new data: predict labels for unknown data using predict() method
    y_pred = model.predict(X_test)
    
    mae= mean_absolute_error(y_test, y_pred)
    maes.append(mae)
    print(f'Absolute Error on prediction for {year} = {mae}')
    
    
```

```python
#Leave-One-Out Cross-Validation Mean Absolute Error
pd.Series(maes).mean()
```
